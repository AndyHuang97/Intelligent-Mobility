## Save distance matrix

```python
%%bash
source activate scipydev
#cd /content/scipy
cd /content/drive/MyDrive/MIDA2/IntelligentMobilityProject/Code/scipy

python
# --- import data
import pandas as pd
import numpy as np
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage

def wheom_prepare_w(w):
  w[w>1] = 0.5 * w[w>1]/(w[w>1]-1)
  return w

def vdm_prepare_df(df, cat_ix):
  replace_arr = {}
  for col in df.columns[~cat_ix]:
    replace_arr[col] = { value: key for key, value in enumerate(np.sort(df[col].unique()))}
  return df.replace(replace_arr)


target_class = "binary"
metric="VDM"
std = True
if metric != "HVDM":
  std = False
if_std = '_std' if std else ''
vars = 5
regions = False
if vars != 7:
  regions = False
if_regions = '_regions' if regions else ''
HC_base_path = "/content/drive/MyDrive/MIDA2/IntelligentMobilityProject/Code/Data/HC/"
algo = "HC"
file_type = ".npy"
ds_path = f'/content/drive/MyDrive/MIDA2/IntelligentMobilityProject/Survey+dataset/4_DNA_{vars}values_normalized.csv'
df = pd.read_csv(ds_path)#.drop("Gender", axis=1)
w = np.nanmax(df, axis = 0) - np.nanmin(df, axis = 0)
cat_ix = np.array([i in range(3, len(df.columns)) for i in range(len(df.columns))])
# WHEOM
# the weight is the number of categories when the variable is categorical
if metric == "WHEOM":
  w[cat_ix] = np.array([len(np.unique(df[col])) for col in df.columns])[cat_ix]
  w = wheom_prepare_w(w)
elif metric == "HEOM":
  w[cat_ix] = 0
elif metric == "HVDM":
  w = None
elif metric == "VDM":
  w = None
  df = vdm_prepare_df(df, cat_ix)

distance_matrix = squareform(pdist(df, metric=metric, w=w))

print(distance_matrix/vars)
print(distance_matrix.shape)
np.save(f"/content/drive/MyDrive/MIDA2/IntelligentMobilityProject/Code/Data/distance_matrix_" \
        f"{metric}{vars}{if_regions}{if_std}_{target_class}.npy", distance_matrix)


#linkages = ["single", "complete", "average", "weighted"]# "ward", "centroid", "median"]
#for link in linkages:
#  HC_path = HC_base_path + metric + str(vars) + '/' + algo + sep + metric + sep + link + file_type
#
#  Z = {}
#  Z[metric] = linkage(df, link, metric=metric)
#  np.save(HC_path, Z[metric], allow_pickle=False)
#  print(link, "done")
```

# Plot TSNE

```python
import numpy as np
import pandas as pd

HC_path = '/content/drive/MyDrive/MIDA2/IntelligentMobilityProject/Code/Data/HC/'
ds_path = ds_path = '/content/drive/MyDrive/MIDA2/IntelligentMobilityProject/Survey+dataset/4_DNA_values.csv'
df_clusters = pd.read_csv(ds_path)
df = df_clusters.copy().drop("cluster", axis=1)

df_DNA = pd.read_csv(ds_path)
clusters = df_clusters["cluster"].values
#clusters = np.load(HC_path+"HC_complete_heom_3clusters.npy")

df_dict = {}
for i in df_DNA.columns:
  df_dict[i] = df_DNA[i].values
```

```python
def compile_tsne_df(tsne_result, df_dict, clusters, n_components=2):
  if n_components == 2:
    components = {'x':tsne_result[:,0], 'y':tsne_result[:,1]}
  else:
    components = {'x':tsne_result[:,0], 'y':tsne_result[:,1], 'z':tsne_result[:,2]}

  df_tsne = pd.DataFrame({**components, **df_dict})
  df_tsne["cluster"] = clusters
  return df_tsne
```

```python
import plotly.express as px

def plot_tsne_updated(df, column, perp, niter, dims):
  df["cluster"] = df["cluster"].map(str)
  title = f"Visualizing Clusters in {dims} Dimensions Using T-SNE with perplexity {perp}"
  if dims==2:
    fig = px.scatter(df, x="x", y="y", color="cluster", title=title)
  else:
    fig = px.scatter_3d(df, x="x", y="y", z="z", color=column, title=title)
  fig.show()

  TSNE_path = "/content/drive/MyDrive/MIDA2/IntelligentMobilityProject/Pictures/DNA_values_v1/TSNE/"
  save_path = f"{TSNE_path}TSNE_perp{perp}_comp{dims}_niter{niter}.html"
  fig.write_html(save_path)
```

```python
for perp in [50]:
  niter=10000
  n_components = 2
  data_path = f"/content/drive/MyDrive/MIDA2/IntelligentMobilityProject/Code/Data/DNA_values_v1/TSNE/postprocessing/tsne_perp{perp}_comp{n_components}_niter{niter}.npy"
  tsne_result = np.load(data_path)
  df_tsne = compile_tsne_df(tsne_result, df_dict, clusters, n_components)
  column = "cluster"
  dims = 2

  plot_tsne_updated(df_tsne, column, perp, niter, dims)
```
