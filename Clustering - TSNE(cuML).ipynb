# TSNE

```python
from cuml.manifold import TSNE

perplexity=40
n_components=2
n_iter=100000
metric="euclidean"
tsne = TSNE(n_components=n_components, 
            verbose=1,
            perplexity=perplexity, 
            random_state=2867976,
            metric=metric,
            n_iter=n_iter,
            )
tsne_result = tsne.fit_transform(df_DNA)
data_path = "/content/drive/MyDrive/MIDA2/IntelligentMobilityProject/Code/Data/TSNE/cuML/tsne_" + metric + "_perp"+str(perplexity)+"_comp"+str(n_components)+"_niter"+str(n_iter)+".npy"
np.save(data_path, tsne_result)
```
